인덱스 변경하기

```python
df_temp.set_index('연도', inplace = True)
df_temp
```

데이터의 선 그래프

```python
plt.plot(df_temp.index , df_temp['평균기온'])
plt.show()
```

그래프의 최저,최고 기온 표시하기

```python
plt.plot(df_temp.index , df_temp['최저기온'])
plt.plot(df_temp.index , df_temp['최고기온'])
plt.show
```

그래프 제목표시하기

```python
plt.plot(df_temp.index, df_temp['평균기온'])
plt.title('avg temp')
plt.show()
plt.plot(df_temp.index, df_temp['최고기온'])
plt.title('max temp')
plt.show()
```

평균기온, 최저기온, 최고기온 한 차트 만들기

```python
plt.plot(df_temp.index, df_temp['평균기온'])
plt.plot(df_temp.index, df_temp['최저기온'])
plt.plot(df_temp.index, df_temp['최고기온'])
plt.title('seoul temp')#한글로 하면 □로 뜹니다.

plt.show()
```

가로 라벨, 세로 라벨 표시

```python
plt.plot(df_temp.index, df_temp['평균기온'])
plt.plot(df_temp.index, df_temp['최저기온'])
plt.plot(df_temp.index, df_temp['최고기온'])
plt.title('seoul temp')#한글로 하면 □로 뜹니다.
plt.xlabel('Years')
plt.ylabel('°c')
plt.show()
```

그래프의 차트 표시하기

```python
plt.plot(df_temp.index, df_temp['평균기온'],label = 'average')
plt.plot(df_temp.index, df_temp['최저기온'],label = 'lowest')
plt.plot(df_temp.index, df_temp['최고기온'],label = 'best')
plt.title('seoul temp')#한글로 하면 □로 뜹니다.
plt.xlabel('Years')
plt.ylabel('°c')
plt.legend()
plt.show()
```

차트 색 바꾸기

```python
plt.plot(df_temp.index, df_temp['평균기온'],label = 'average', color = 'green')
plt.plot(df_temp.index, df_temp['최저기온'],label = 'lowest', color = 'blue')
plt.plot(df_temp.index, df_temp['최고기온'],label = 'best', color = 'black')
plt.title('seoul temp')#한글로 하면 □로 뜹니다.
plt.xlabel('Years')
plt.ylabel('°c')
plt.legend()
plt.show()
```

마커 표시하기(x)

```python
plt.plot(df_temp.index, df_temp['평균기온'],label = 'average', color = 'green',marker = 'X')
plt.plot(df_temp.index, df_temp['최저기온'],label = 'lowest', color = 'blue',marker = 'X')
plt.plot(df_temp.index, df_temp['최고기온'],label = 'best', color = 'black',marker = 'X')
plt.title('seoul temp')#한글로 하면 □로 뜹니다.
plt.xlabel('Years')
plt.ylabel('°c')
plt.legend()
plt.show()
```

데이터 크기

```python
plt.figure(figsize = (11, 7))
plt.plot(df_temp.index, df_temp['평균기온'],label = 'average', color = 'green',marker = 'X')
plt.plot(df_temp.index, df_temp['최저기온'],label = 'lowest', color = 'blue',marker = 'X')
plt.plot(df_temp.index, df_temp['최고기온'],label = 'best', color = 'black',marker = 'X')
plt.title('seoul temp')#한글로 하면 □로 뜹니다.
plt.xlabel('Years')
plt.ylabel('°c')
plt.legend()
plt.show()
```
